```python
import pandas as pd
import Simulation as sim

import matplotlib.pyplot as plt
intervals = [25, 50, 100, 150, 200]
intervas2 = [1,5,10]

for i in intervals:
    result = sim.main(i, 100, 3, 1)
    result = pd.DataFrame(result)

    plt.scatter(result['ProcessNumber'], result['time'])
    plt.xlabel('Numero De Proceso')
    plt.ylabel('Tiempo')
    plt.axvspan(0, 0, color='red', alpha=0.2, label=f'Media: {result["time"].mean()}')
    plt.axhspan(0, 0, color='green', alpha=0.2, label=f'Desviacion: {result["time"].std()}')
    plt.legend()
    plt.show()
```
